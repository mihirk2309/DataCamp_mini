```python
plt.scatter(df["Electricity(kWh)"], df["TotalGHGEmissions"],  color='blue')
plt.xlabel("Electricity(kWh)")
plt.ylabel("TotalGHGEmissions")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(df[['Electricity(kWh)']])
train_y = np.asanyarray(df[['TotalGHGEmissions']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(df["Electricity(kWh)"], df["TotalGHGEmissions"],  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Electricity(kWh)")
plt.ylabel("TotalGHGEmissions")
plt.show()
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(df[['Electricity(kWh)']])
test_y = np.asanyarray(df[['TotalGHGEmissions']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```
